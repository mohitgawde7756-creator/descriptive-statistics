 4. : Median in Real Life
A car dealer notes the prices of used cars: [$8,000, $9,500, $10,200, $11,000, $50,000].
Why is the median a better measure than the mean in this case? Calculate the median.
-    Calculate the Median

Given prices (already in order):
[$8,000, $9,500, $10,200, $11,000, $50,000]

The middle value is $10,200
 Final Answer:

Median = $10,200

 Why is median better than mean here?
The price $50,000 is an extreme value (outlier) compared to the others.
If we calculate the mean, this high value would pull the average up, making it misleading.
The median is not affected by outliers, so it shows the true typical price of most cars.

 Conclusion:
The median ($10,200) gives a more accurate idea of what a typical used car costs, while the mean would be distorted by the unusually expensive $50,000 car.

7.  Variance for Decision-Making (Medium)
Two delivery companies track delivery delays (in minutes).
Company A: variance = 6
Company B: variance = 15
Which company is more consistent, and why?
-    Final Answer:

Company A is more consistent.

 Why?
Variance measures how spread out the data is (how much delays vary from the average).
Lower variance = more consistency
Higher variance = more fluctuation
 Compare the two:
Company A → Variance = 6 (less spread, more stable delays)
Company B → Variance = 15 (more spread, less predictable delays)
 Conclusion:

Company A has more consistent delivery times because its delays vary less.
This makes it more reliable for customers compared to Company B.

8.  Standard Deviation in Context
A finance student compares the daily price fluctuations of two cryptocurrencies.
Coin A: standard deviation = $30
Coin B: standard deviation = $120
Which coin is riskier to invest in, and why?
-    Final Answer:

Coin B is riskier to invest in.

 Why?
Standard deviation measures volatility (how much prices fluctuate).
Higher standard deviation = more price swings = higher risk
Lower standard deviation = more stable prices = lower risk
 Compare the two:
Coin A → Standard Deviation = $30 (more stable)
Coin B → Standard Deviation = $120 (highly volatile)
 Explanation:
Coin B’s price changes are much larger and unpredictable.
This means:
 Potential for higher profit
 But also higher chance of loss
 Conclusion:

Coin B is riskier because its price fluctuates much more.
Investors who prefer stability would choose Coin A, while risk-takers might consider Coin B for higher returns.
